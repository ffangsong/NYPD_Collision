```python
df_2017_area = df_2017[(df['LATITUDE']>35) & (df['LONGITUDE']>-80) & (df['LONGITUDE']< -60)]
df_2017_area.info();
df_2017_area.head()
```

```python
df_2017_byZipCode = df_2017_area.groupby('ZIP CODE')
df_2017_AreaStd= df_2017_byZipCode.std()[['LONGITUDE','LATITUDE']]
df_2017_AreaStd
```

```python
df_2017_TotalCollision = df_2017_area.groupby('ZIP CODE').count()[['DATE']]
#df_2017_TotalCollision['ZIP CODE'] = df_2017_TotalCollision['ZIP CODE'].map(np.int64)
df_2017_TotalCollision
```

```python
df_2017_CollisonPerArea = pd.merge(df_2017_TotalCollision,df_2017_AreaStd,left_index=True,right_index =True)
df_2017_CollisonPerArea
```

```python
df_2017_CollisonPerArea_1000=df_2017_CollisonPerArea[df_2017_CollisonPerArea['DATE']>1000]
```

```python
import math
```

```python
df_2017_CollisonPerArea_1000['AREA'] = math.pi *df_2017_CollisonPerArea_1000['LONGITUDE']*111*0.25945998191 * df_2017_CollisonPerArea_1000['LATITUDE']*111
df_2017_CollisonPerArea_1000
```

```python
df_2017_CollisonPerArea_1000['CollisionPerArea'] = df_2017_CollisonPerArea_1000['DATE']/df_2017_CollisonPerArea_1000['AREA']
```
